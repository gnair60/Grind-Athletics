```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
from bs4 import BeautifulSoup
import csv

# Selenium to get High School data
def extract_high_schools():
    driver = webdriver.Chrome()
    driver.get("https://utahutes.com/sports/mens-basketball/roster")

    # Find the high school details dynamically loaded
    players = driver.find_elements(By.CLASS_NAME, "sidearm-roster-player-highschool")

    # Extract and clean high school details
    high_schools = [player.text.strip() for player in players if player.text.strip()]

    driver.quit()
    return high_schools

# BeautifulSoup to get other player details
def extract_player_details():
    # URL of the roster page
    url = "https://utahutes.com/sports/mens-basketball/roster"

    # Send a GET request to the webpage
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all player containers
        players = soup.find_all('div', class_='sidearm-list-card-details relative')

        # Keywords to identify non-player roles
        non_player_keywords = [
            "Coach", "General Manager", "Director", "Trainer", "Performance", "Assistant"
        ]

        # Initialize a list to store player details
        player_data = []

        for player in players:
            # Check if the role indicates non-player staff
            role_element = player.find('div', class_='sidearm-list-card-details-item')
            if role_element and any(keyword in role_element.text for keyword in non_player_keywords):
                # Skip this entry if it matches any non-player keyword
                continue

            # Extract the player name
            name_element = player.find('a', class_='sidearm-roster-player-name')
            if name_element:
                full_name = ' '.join(name_element.text.strip().split())
            else:
                continue  # Skip if no name is found

            # Extract height
            height_element = player.find('span', class_='sidearm-roster-player-height')
            height = height_element.text.strip() if height_element else "N/A"

            # Extract position
            position_element = player.find('div', class_='sidearm-roster-player-position-short')
            position = position_element.text.strip() if position_element else "N/A"

            # Extract academic class
            class_element = player.find('span', class_='sidearm-roster-player-academic-year')
            academic_class = class_element.text.strip() if class_element else "N/A"

            # Extract hometown
            hometown_element = player.find('span', class_='sidearm-roster-player-hometown')
            hometown = hometown_element.text.strip() if hometown_element else "N/A"

            # Extract previous school
            previous_school_element = player.find('span', class_='sidearm-roster-player-previous-school')
            previous_school = previous_school_element.text.strip() if previous_school_element else "N/A"

            # Store all details in a dictionary
            player_data.append({
                "Name": full_name,
                "Height": height,
                "Position": position,
                "Class": academic_class,
                "Hometown": hometown,
                "Previous School": previous_school
            })

        return player_data
    else:
        print(f"Failed to fetch the webpage. Status code: {response.status_code}")
        return []

# Combine data and save to CSV
def combine_data_and_save():
    high_schools = extract_high_schools()
    player_details = extract_player_details()

    # Ensure lengths match
    if len(player_details) != len(high_schools):
        print("Mismatch in player details and high school data lengths.")
        return

    # Add high schools to player details
    for i, player in enumerate(player_details):
        player['High School'] = high_schools[i] if i < len(high_schools) else "N/A"

    # Save the data to a CSV file
    with open('final_utah_utes_player_details.csv', 'w', newline='', encoding='utf-8') as csvfile:
        csvwriter = csv.writer(csvfile)
        # Write the header row
        csvwriter.writerow(['Name', 'Height', 'Position', 'Class', 'Hometown', 'High School', 'Previous School'])
        # Write player data
        for data in player_details:
            csvwriter.writerow([
                data['Name'], data['Height'], data['Position'], data['Class'],
                data['Hometown'], data['High School'], data['Previous School']
            ])

    print("Player data successfully saved to 'final_utah_utes_player_details.csv'")

# Run the script
combine_data_and_save()
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import unicodedata

# Base URL of the BYU roster page
base_url = "https://byucougars.com/sports/womens-basketball/roster/season/"

# Range of years to scrape
start_year = 2023
end_year = 2024

# Function to format height
def format_height(height_str):
    try:
        # Ensure the string is numeric and has the expected length
        if height_str.isdigit() and len(height_str) > 2:
            feet = height_str[:-2]  # All but the last two digits
            inches = height_str[-2:]  # The last two digits
            return f"{int(feet)}'{int(inches)}\""
        elif height_str.isdigit():
            return f"{int(height_str)}'"
    except ValueError as e:
        print(f"Error formatting height: {height_str} -> {e}")
    return "N/A"

# CSV file to save the data
csv_file = "Byu_Womens.csv"

# Open the CSV file for writing
with open(csv_file, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    # Write the header row
    writer.writerow(["Season", "Name", "Height", "Class Year", "Hometown", "Previous School"])

    # Iterate through each season
    for year in range(start_year, end_year + 1):
        # Construct the URL for the current year
        url = f"{base_url}{year}-{year + 1}"

        # Send a GET request to the webpage
        response = requests.get(url)

        # Check if the request was successful
        if response.status_code == 200:
            print(f"Fetching data for the season {year}-{year + 1}...")
            # Parse the HTML content
            soup = BeautifulSoup(response.text, 'html.parser')

            # Find the main player roster container
            player_section = soup.find('div', class_='roster-players-cards')
            if not player_section:
                print(f"No player section found for {year}-{year + 1}.")
                continue

            # Find all player cards
            player_cards = player_section.find_all('div', class_='roster-card-item')

            # Check if player cards exist
            if not player_cards:
                print(f"No player data found for {year}-{year + 1}.")
                continue

            # Loop through each player card to extract details
            for card in player_cards:
                # Extract player name
                name_element = card.find('a', class_='roster-card-item__title-link')
                name = name_element.text.strip() if name_element else "N/A"

                # Extract height and class year
                basic_fields = card.find_all('span', class_='roster-player-card-profile-field__value--basic')
                height = basic_fields[0].text.strip() if len(basic_fields) > 0 else "N/A"
                class_year = basic_fields[1].text.strip() if len(basic_fields) > 1 else "N/A"

                # Debugging: Print the extracted raw height

                # Normalize and format the height
                height = format_height(unicodedata.normalize("NFKD", height).encode("ascii", "ignore").decode("utf-8"))

                # Extract hometown
                hometown_element = card.find('span', class_='roster-player-card-profile-field__value--hometown')
                hometown = hometown_element.text.strip() if hometown_element else "N/A"

                # Extract previous school
                previous_school_element = card.find('span', class_='roster-player-card-profile-field__value--previous-school')
                previous_school = previous_school_element.text.strip() if previous_school_element else "N/A"

                # Write the player's details to the CSV file
                writer.writerow([f"{year}-{year + 1}", name, height, class_year, hometown, previous_school])

        else:
            print(f"Failed to fetch data for {year}-{year + 1}. Status code: {response.status_code}")

print(f"Data has been saved to {csv_file}.")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

# Base URL for Southern Utah Men's Basketball roster
base_url = "https://suutbirds.com/sports/mens-basketball/roster/"

# Range of years to scrape
start_year = 2020
end_year = 2024

# Open a CSV file to save data
csv_file = 'Southern_Utah_Mens.csv'
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Season', 'Name', 'Position', 'Height', 'Academic Year', 'Hometown', 'High School','Previous School'])

    # Iterate through each season and scrape data
    for year in range(start_year, end_year + 1):
        # Construct the URL for the season
        url = f"{base_url}{year}-{str(year + 1)[-2:]}"  # Example: /2020-21, /2023-24
        print(f"\nFetching data for the {year}-{year + 1} season... URL: {url}")

        # Send a GET request to the webpage
        response = requests.get(url)

        # Check if the request was successful
        if response.status_code == 200:
            # Parse the HTML content
            soup = BeautifulSoup(response.text, 'html.parser')

            # Find all roster cards
            roster_cards = soup.find_all('li', class_='sidearm-list-card-item')

            # Check if roster cards exist
            if not roster_cards:
                print(f"No player data found for the {year}-{year + 1} season.")
                continue

            # Loop through each roster card
            for card in roster_cards:
                try:
                    # Skip cards with "Coach" or "Staff" in the title or description
                    if "Coach" in card.text or "Staff" in card.text or "Assistant" in card.text or "Trainer" in card.text:
                        continue

                    # Extract player name
                    first_name = card.find('span', class_='sidearm-roster-player-first-name')
                    last_name = card.find('span', class_='sidearm-roster-player-last-name')
                    name = f"{first_name.text.strip()} {last_name.text.strip()}" if first_name and last_name else "N/A"

                    # Extract position
                    position_element = card.find('div', class_='sidearm-list-card-details-item sidearm-roster-player-position-short')
                    position = position_element.text.strip() if position_element else "N/A"

                    # Extract height
                    height_element = card.find('span', class_='sidearm-roster-player-height')
                    height = height_element.text.strip() if height_element else "N/A"

                    # Extract academic year
                    academic_year_element = card.find('span', class_='sidearm-roster-player-academic-year')
                    academic_year = academic_year_element.text.strip() if academic_year_element else "N/A"

                    # Extract hometown
                    hometown_element = card.find('span', class_='sidearm-roster-player-hometown')
                    hometown = hometown_element.text.strip() if hometown_element else "N/A"
                    
                    high_school_element = card.find('span', class_='sidearm-roster-player-highschool')
                    high_school = high_school_element.text.strip() if high_school_element else "N/A"

                    # Extract previous school
                    previous_school_element = card.find('span', class_='sidearm-roster-player-previous-school')
                    previous_school = previous_school_element.text.strip() if previous_school_element else "N/A"

                    # Write to CSV
                    writer.writerow([f"{year}-{year + 1}", name, position, height, academic_year, hometown, previous_school])

                    # Print the player's details
                    print(f"Season: {year}-{year + 1}, Name: {name}, Position: {position}, Height: {height}, "
                          f"Academic Year: {academic_year}, Hometown: {hometown}, High School: {high_school}, Previous School: {previous_school}")

                except Exception as e:
                    print(f"Error extracting data for a player: {e}")
        else:
            print(f"Failed to fetch data for {year}-{year + 1}. Status code: {response.status_code}")

print(f"\nData has been saved to {csv_file}.")

```

```python
import requests
from bs4 import BeautifulSoup
import csv

# Base URL for Salt Lake Community College Men's Basketball roster
base_url = "https://slccbruins.com/sports/mens-basketball/roster/"

# Range of years to scrape
start_year = 2020
end_year = 2024

# File to save data
output_file = "Salt_Lake_Community_Mens.csv"

# Prepare the CSV file
with open(output_file, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    # Write the header
    writer.writerow(["Season", "Name", "Position", "Height", "Academic Year", "Hometown", "High School"])

    for year in range(start_year, end_year + 1):
        # Construct the URL for the season
        url = f"{base_url}{year}-{str(year + 1)[-2:]}"  # Example: /2020-21, /2023-24
        print(f"\nFetching player details for the {year}-{year + 1} season... URL: {url}")

        # Send a GET request to the webpage
        response = requests.get(url)

        if response.status_code == 200:
            # Parse the HTML content
            soup = BeautifulSoup(response.text, "html.parser")

            # Find all player cards
            player_cards = soup.find_all("li", class_="sidearm-roster-player")

            if not player_cards:
                print(f"No player data found for the {year}-{year + 1} season.")
                continue

            # Loop through each player card to extract details
            for card in player_cards:
                try:
                    # Extract full name (first and last name)
                    name_container = card.find("h3")
                    if name_container:
                        name = name_container.get_text(strip=True)
                    else:
                        name = "N/A"

                    # Extract position
                    position_element = card.find("div", class_="sidearm-roster-player-position")
                    if position_element:
                        position_text = position_element.find("span", class_="text-bold")
                        position = position_text.text.strip() if position_text else "N/A"
                    else:
                        position = "N/A"

                    # Extract height
                    height_element = card.find("span", class_="sidearm-roster-player-height")
                    height = height_element.text.strip() if height_element else "N/A"

                    # Extract academic year
                    academic_year_element = card.find("span", class_="sidearm-roster-player-academic-year")
                    academic_year = academic_year_element.text.strip() if academic_year_element else "N/A"

                    # Extract hometown
                    hometown_element = card.find("span", class_="sidearm-roster-player-hometown")
                    hometown = hometown_element.text.strip() if hometown_element else "N/A"

                    # Extract high school
                    high_school_element = card.find("span", class_="sidearm-roster-player-highschool")
                    high_school = high_school_element.text.strip() if high_school_element else "N/A"

                    # Write player details to the CSV file
                    writer.writerow([f"{year}-{year + 1}", name, position, height, academic_year, hometown, high_school])

                    # Print player details to the console
                    print(f"Season: {year}-{year + 1}, Name: {name}, Position: {position}, Height: {height}, "
                          f"Academic Year: {academic_year}, Hometown: {hometown}, High School: {high_school}")

                except Exception as e:
                    print(f"Error extracting data for a player: {e}")
        else:
            print(f"Failed to fetch data for the {year}-{year + 1} season. Status code: {response.status_code}")

print(f"\nData saved to {output_file}")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

# Base URL for Utah State Eastern Men's Basketball roster
base_url = "https://usueasternathletics.com/sports/mens-basketball/roster"

# Fetch and process the roster data
def fetch_roster_data(url, season, data):
    print(f"\nFetching player details for the {season} season... URL: {url}")
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Locate all player cards
        player_cards = soup.find_all('li', class_='sidearm-roster-player')

        for card in player_cards:
            try:
                # Extract player name
                name_tag = card.find('a', attrs={'aria-label': True})
                name = name_tag['aria-label'].split('-')[0].strip() if name_tag else "N/A"

                # Extract position
                position_tag = card.find('span', class_='text-bold')
                position = position_tag.text.strip() if position_tag else "N/A"

                # Extract height
                height_tag = card.find('span', class_='sidearm-roster-player-height')
                height = height_tag.text.strip() if height_tag else "N/A"

                # Extract academic year
                academic_year_tag = card.find('span', class_='sidearm-roster-player-academic-year')
                academic_year = academic_year_tag.text.strip() if academic_year_tag else "N/A"

                # Extract hometown
                hometown_tag = card.find('span', class_='sidearm-roster-player-hometown')
                hometown = hometown_tag.text.strip() if hometown_tag else "N/A"

                # Extract high school
                high_school_tag = card.find('span', class_='sidearm-roster-player-highschool')
                high_school = high_school_tag.text.strip() if high_school_tag else "N/A"

                # Append the player's details to the data list
                data.append([season, name, position, height, academic_year, hometown, high_school])

            except Exception as e:
                print(f"Error processing a player card: {e}")
    else:
        print(f"Failed to fetch data for the {season} season. Status code: {response.status_code}")

# Main execution
if __name__ == "__main__":
    # Define the CSV file name
    csv_file = "Utah_State_Eagles_Mens.csv"
    # Header for the CSV file
    headers = ["Season", "Name", "Position", "Height", "Academic Year", "Hometown", "High School"]
    # List to store player data
    player_data = []

    for year in range(2021, 2025):  # Loop through seasons 2021-2025
        season = f"{year}-{year + 1}"
        url = f"{base_url}/{year}-{year + 1}"  # Add year range to URL
        fetch_roster_data(url, season, player_data)

    # Write data to CSV
    with open(csv_file, mode="w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        # Write the header
        writer.writerow(headers)
        # Write the player data
        writer.writerows(player_data)

    print(f"\nData saved to {csv_file} successfully!")
```
