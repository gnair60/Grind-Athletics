```python
import requests
from bs4 import BeautifulSoup

def get_player_profiles(base_search_url, player_name, state, sport, school_keyword):
    """
    Fetches all player profile links for the given player name, state, sport,
    and filters based on the school keyword.
    """
    # Format search URL
    search_url = f"{base_search_url}?q={player_name.replace(' ', '+')}&state={state}&sport={sport}"
    print(f"Searching for player profiles: {search_url}")

    response = requests.get(search_url)
    if response.status_code != 200:
        print(f"Failed to load search page. Status code: {response.status_code}")
        return []

    soup = BeautifulSoup(response.text, 'html.parser')
    player_links = soup.find_all('a', href=True, title=True)
    print(f"Found {len(player_links)} potential player links.")

    profiles = []
    for link in player_links:
        href = link.get('href', '')
        title = link.get('title', '')
        if "/athletes/" in href:  # Filter player profile links
            profile_url = "https://www.maxpreps.com" + href
            # Check if the school_keyword is in the URL or title
            if school_keyword.lower() in href.lower() or school_keyword.lower() in title.lower():
                profiles.append(profile_url)
                print(f"Matched Player Profile: {title} -> {profile_url}")

    return profiles


def validate_player_profile(profile_url):
    """
    Checks if a player profile URL is valid and accessible.
    """
    response = requests.get(profile_url)
    if response.status_code == 200:
        print(f"Valid Player Profile: {profile_url}")
        return True
    else:
        print(f"Invalid Player Profile: {profile_url}")
        return False


def get_stats_link(profile_url):
    """
    Fetches the Stats section link from a player's profile if available.
    """
    response = requests.get(profile_url)
    if response.status_code != 200:
        print(f"Failed to load player profile. Status code: {response.status_code}")
        return None

    soup = BeautifulSoup(response.text, 'html.parser')
    stats_link = soup.find('a', string='Stats')

    if stats_link:
        stats_url = "https://www.maxpreps.com" + stats_link['href']
        print(f"Stats URL Found: {stats_url}")
        return stats_url
    else:
        print(f"No Stats URL found for profile: {profile_url}")
        return None


def find_valid_stats_url(base_search_url, player_name, state, sport, school_keyword):
    """
    Searches for a player's valid stats URL by iterating over all possible profiles
    and filtering using the school keyword.
    """
    player_profiles = get_player_profiles(base_search_url, player_name, state, sport, school_keyword)

    if not player_profiles:
        print("No player profiles found.")
        return None

    for profile_url in player_profiles:
        if validate_player_profile(profile_url):
            stats_url = get_stats_link(profile_url)
            if stats_url:  # Return the first valid stats URL
                return stats_url

    print("No valid stats URL found among the profiles.")
    return None


# Example Usage
base_search_url = "https://www.maxpreps.com/search/"
player_name = "Hunter Erickson"
state = "ut"
sport = "basketball"
school_keyword = "timpview"  # Part of the school name to filter the correct profile

# Get the valid stats URL for the player
stats_url = find_valid_stats_url(base_search_url, player_name, state, sport, school_keyword)

if stats_url:
    print(f"Final Stats URL: {stats_url}")
else:
    print("Unable to find a valid stats URL.")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time


def fetch_shooting_stats(stats_url):
    """Fetch the shooting stats (PTS, FGM, FGA) for a player from the given URL."""
    # Configure ChromeDriver
    options = Options()
    options.headless = True  # Set to False if you want to see the browser
    driver = webdriver.Chrome(options=options)

    try:
        # Navigate to the player's stats page
        driver.get(stats_url)

        # Allow time for the page to load
        time.sleep(5)

        # Click on the "Shooting" tab
        shooting_tab = driver.find_element(By.XPATH, '//button[@title="Shooting"]')
        shooting_tab.click()
        time.sleep(3)  # Wait for the table to load

        # Locate the shooting stats table rows
        rows = driver.find_elements(By.XPATH, '//tbody[contains(@class, "table-body")]/tr')

        # Debug: Check if rows are located
        if not rows:
            raise Exception("No rows found in the shooting stats table!")

        # Extract stats
        shooting_stats = []
        for row in rows:
            cells = row.find_elements(By.TAG_NAME, 'td')
            try:
                if len(cells) > 4:  # Ensure there are enough cells
                    pts = cells[2].text.strip()  # Points
                    fgm = cells[3].text.strip()  # Field Goals Made
                    fga = cells[4].text.strip()  # Field Goals Attempted
                    fg = cells[5].text.strip()  # Field Goals Attempted


                    # Append the data as a dictionary
                    shooting_stats.append({
                        'PTS': pts,
                        'FGM': fgm,
                        'FGA': fga,
                        'FG%': fg
                    })
            except Exception as row_error:
                print(f"Error parsing row: {row.text}, Error: {row_error}")

        # Get the record with the latest year
        if shooting_stats:
            # Assuming rows are sorted by year, take the first one
            latest_shooting_stats = shooting_stats[0]
            return latest_shooting_stats
        else:
            return "No shooting stats found."

    except Exception as e:
        print(f"An error occurred: {e}")
        return None

    finally:
        driver.quit()


# Example usage
stats_url = "https://www.maxpreps.com/co/highlands-ranch/thunderridge-grizzlies/athletes/zach-keller/basketball/stats/?careerid=3b4h6c0jpi0g0"
shooting_stats = fetch_shooting_stats(stats_url)

if shooting_stats:
    print("Shooting Stats (PTS, FGM, FGA, FG%):", shooting_stats)
else:
    print("Failed to fetch shooting stats.")
```
